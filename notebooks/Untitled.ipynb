```python
import matplotlib.pyplot as plt
import numpy as np

# Create a gradient
gradient = np.linspace(0, 1, 256).reshape(1, -1)

# Plot all colormaps
colormaps = plt.colormaps()
n_maps = len(colormaps)
fig, axes = plt.subplots(n_maps // 5 + 1, 5, figsize=(12, 3 * (n_maps // 5 + 1)))
axes = axes.flatten()

for i, cmap_name in enumerate(colormaps):
    axes[i].imshow(gradient, aspect='auto', cmap=cmap_name)
    axes[i].set_title(cmap_name, fontsize=8)
    axes[i].axis('off')

# Turn off any unused axes
for j in range(i + 1, len(axes)):
    axes[j].axis('off')

plt.tight_layout()
plt.show()
```
